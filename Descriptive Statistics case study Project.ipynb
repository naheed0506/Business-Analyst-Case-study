```python
# Create a scatterplot of the age and employee salary features
plt.figure(figsize = (10,5))
sns.scatterplot(data = data, x = 'age', y = 'salary', color = 'red', 
               edgecolor = 'linen', alpha = 0.5)
plt.title("Scatterplott of Employee Age v/s Employee Salary")
plt.xlabel("Employee Age")
plt.ylabel("Employee Salary")
plt.show()
```

```python
# Create a scatterplot of the healthy_eating and active_lifestyle features
plt.figure(figsize = (10,5))
sns.scatterplot(data = data, x = 'healthy_eating', y = 'active_lifestyle', color = 'blue', 
               edgecolor = 'linen', alpha = 0.5)
plt.title("Scatterplott of Healthy Eating v/s Active Lifestyle")
plt.xlabel("Healthy Eating")
plt.ylabel("Active Lifestyle")
plt.show()
```

```python
# Create a scatterplot of the healthy_eating and salary features
plt.figure(figsize = (10,5))
sns.scatterplot(data = data, x = 'healthy_eating', y = 'salary', color = 'green', 
               edgecolor = 'linen', alpha = 0.5)
plt.title("Scatterplott of Healthy Eating v/s Salary")
plt.xlabel("Healthy Eating")
plt.ylabel("Salary")
plt.show()
```

```python
# create a count lot of the groups features
plt.figure(figsize = (10,5))
sns.countplot(data = data, x = "groups", edgecolor = "linen", alpha = 0.7, color = "red")
plt.title("Count Plot of Groups")
plt.xlabel("Groups")
plt.ylabel("Count")
plt.show()
```

```python
# Create a histogram of the salary
plt.figure( figsize = (11,6))
sns.histplot(data = data, x = "salary", color = "orange", edgecolor = "linen", alpha = 0.7, bins = 10)
plt.title("Histogram of Salary")
plt.xlabel("Salary")
plt.ylabel("Count")
plt.show()
```
