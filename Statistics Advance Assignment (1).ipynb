```python
# Let's tackle these features: "Blue", "Wi_Fi", "Tch_Scr", "Ext_Mem"
df = pd.DataFrame(data, columns = ['Blue', 'Wi_Fi', 'Tch_Scr', 'Ext_Mem'])
df
```

```python
# The children want phones that have the following: Bluetooth, WiFi, touch screen and external memory support
# Create a logical condition for this situation and store the logical values as "con
con1 = (df["Blue"] == 1) & (df["Wi_Fi"] == 1) & (df["Tch_Scr"] == 1) & (df["Ext_Mem"] == 1)
```

```python
con1
```

```python
# Let's tackle these features: "Px_h", "Px_w"
df6 = pd.DataFrame(data, columns = ['Px_h', 'Px_w'])
df6
```

```python
# Create a new feature called "Px" which stores the total resolution of the screen
df1['Px'] = df1['Px_h']*df1['Px_w']
df1.head()  
```

```python
df1['Px'].mean()
```

```python
df1['Px'].median()
```

```python
# Create a histogram of the "Px" feature and also show the mean and the median
plt.figure(figsize = (11,6))
sns.histplot(data = df1, x = 'Px', color = 'blue', edgecolor = 'linen', alpha = 0.5, bins = 5)
plt.title('Histogram of Px')
plt.xlabel('Width')
plt.ylabel('Height')
plt.vlines(df1['Px'].mean(), ymin = 0, ymax = 25, color = 'yellow', label = 'Mean')
plt.vlines(df1['Px'].median(), ymin = 0, ymax = 25, color = 'orange', label = 'Median')
plt.legend()
plt.show()
```

```python
# The children want phones that have good screen resolutions
# Consider the phones that have screen resolutions greater than or equal to the median value in the data set
# Create a logical condition for this situation and store the logical values as "con2"
df1['Px_h'].mean()
a = df1['Px_h'].median()
a
```

```python
df1['Px_w'].mean()
b = df1['Px_w'].median()
b
```

```python
con2 = df1[(df1['Px_h'] >= a) & (df1["Px_w"] >= b)]
con2
```

```python
# Let's tackle these features: "Scr_h", "Scr_w"
df2 = pd.DataFrame(data, columns = ['Scr_h', 'Scr_w'])
df2
```

```python
# Create a new feature called "Scr_d" which stores the length of the diagonal of the screen of the phone
df2['Scr_d'] = np.sqrt((df2['Scr_h']**2) + (df2['Scr_w']**2))
df2.head(50)
```

```python
df2['Scr_d'].mean()
```

```python
df2['Scr_d'].median()
```

```python
Q1 =np.percentile(df2['Scr_d'], 25)
Q3 =np.percentile(df2['Scr_d'], 75)
```

```python
Q1
```

```python
Q3
```

```python
# Create a histogram of the "Scr_d" feature and also show the quartiles
plt.figure(figsize = (11,6))
sns.histplot(data = df2, x = 'Scr_d', color = 'green', edgecolor = 'linen', alpha = 0.5, bins = 5)
plt.title('Histogram of Scr_d')
plt.xlabel('Scr_d')
plt.ylabel('Count')
plt.vlines(df2['Scr_d'].mean(), ymin = 0, ymax = 40, color = 'red', label = 'Upper')
plt.vlines(df2['Scr_d'].median(), ymin = 0, ymax = 40, color = 'brown', label = 'Lower')
plt.legend()
plt.show()
```

```python
# The children want phones that have very good screen sizes
# Consider the phones that have screen sizes greater than or equal to the upper quartile value in the data set
# Create a logical condition for this situation and store the logical values as "con3"
con3 = df2['Scr_d'] >= Q3
con3
```
